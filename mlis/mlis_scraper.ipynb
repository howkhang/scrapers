```python
import requests
```

```python
s = requests.Session() # create session and reuse
url = 'https://www.mlis.gov.mm/mLsSc.do'
r = s.get(url, verify=False)
r.status_code
```

```python
# pageIndex is used to load the next page
# queryType = 07 is for Translation Law(Name of Law)
url = 'https://www.mlis.gov.mm/mLsScList.do'
payload = {"pageSize":50,"pageIndex":1,"upperLawordKndCode":"0100","queryType":"07","ordrType":"01","query":"","selFont":"PYI"}
r = s.post(url, json=payload)
r.status_code
```

```python
r.json()
```

```python
r.json()['list'][0]
```

```python
r.json()['list'][0]['lawordSn']
```

```python
url2 = 'https://www.mlis.gov.mm/mLsViewDetail.do'
payload = {"lawordSn":r.json()['list'][0]['lawordSn']}
r2 = s.post(url2, json=payload)
r2.status_code
```
